```python
from langchain_core.documents import Document

documents=[
    Document(
        page_content="Dogs are great companions,known for their loyalty and friendliness.", # used for embeddings
        metadata={"source":"mammal-pets-doc"},
    ),
    Document(page_content="cats are independent pets that often enjoy their own space.",
             metadata={"source":"mammal-pets-doc"},),
    Document(page_content="Goldfish are popular pets for beginners ,requiring relatively simple care ",
             metadata={"source":"fish-pets-doc"},),
    Document(page_content="Parrots are intelligent birds capable of mimiking human speech",
             metadata={"source":"birds-pets-doc"},),
    Document(page_content="Rabbits are social animals that need plenty of space to hop around",
             metadata={"source":"mammal-pet-doc"}),
    Document(page_content="Langchain helps building LLM-based applications",
             metadata={"source":"lang-chain-doc"},),
    Document(page_content="LLMs have changed the scene i AI and ML ",
             metadata={"source":"llm-docs"},),
]
```

```python
!pip install langchain-chroma
from langchain_chroma import Chroma
```

```python
!pip install langchain-openai
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
```

```python
vector=Chroma.from_documents(documents,embedding=OpenAIEmbeddings())
```

```python
results=vector.similarity_search("cat",2)
```

```python
for doc in results:
  print(doc.page_content,doc)
  print("-----------")
```

```python
embeddings_model = OpenAIEmbeddings()
query_embedding = embeddings_model.embed_query("cats are cool")
result = vector.similarity_search_by_vector_with_relevance_scores(query_embedding)
```

```python
for doc,score in result:
  print(doc.page_content,score)
  print("----------")
```
